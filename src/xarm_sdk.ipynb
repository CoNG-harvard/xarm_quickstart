# The xArm SDK

```python
from xarm.wrapper import XArmAPI

# Connect to the robot
arm = XArmAPI('192.168.1.230')  # Replace with your robot's IP address

# Enable motion
arm.motion_enable(enable=True)

# Set robot to ready state
arm.set_mode(0)    # Position control mode
arm.set_state(0)   # Set to ready state

# # Move to a target position
current_pos = arm.get_position()[1]
x, y, z, r, p, y = current_pos[0], current_pos[1], current_pos[2], current_pos[3], current_pos[4], current_pos[5]
arm.set_position(x=x+20, y=y+20, z=z, roll=r, pitch=p, yaw=y, speed=100, wait=True)
print(arm.get_position())

# Disconnect
arm.disconnect()
```
